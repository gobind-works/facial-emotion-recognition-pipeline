```python
X_train, X_validation, y_train, y_validation = train_test_split(X_data, y_data, test_size=0.2,random_state=0)
```

```python
n_components = 10
pca =PCA(svd_solver='randomized',n_components=n_components,whiten=True).fit(X_train)

X_train_pca = pca.transform(X_train)

X_val_pca = pca.transform(X_validation)
```

```python
# Assuming `n_classes` is the number of unique classes
class_labels = np.array([f'Emotion {i}' for i in range(n_classes)])  # Replace with actual emotion names if available
class_counts = {label: 0 for label in class_labels}  # Initialize counts

# Count the number of images for each emotion
for class_index in np.argmax(y_train, axis=-1):  # Get the index of the highest probability class
    label = class_labels[class_index]
    class_counts[label] += 1

# Print the number of images per emotion
print("Number of images per emotion:")
for label, count in class_counts.items():
    print(f"{label}: {count}")

# Plot some sample images from each emotion
plt.figure(figsize=(10, 10))
img_shape = 48  # Assuming images are 48x48 pixels

for i in range(7):  # Assuming 7 emotions
    # Get a sample image for the i-th class
    sample_indices = np.where(np.argmax(y_train, axis=-1) == i)[0]  # Indices of images belonging to class i
    if sample_indices.size > 0:  # Check if there are any samples
        img_index = sample_indices[0]  # Get the first image index
        img = X_train[img_index].reshape(img_shape, img_shape)  # Reshape if necessary
        plt.subplot(3, 3, i + 1)
        plt.imshow(img, cmap='gray')  #gray, 'BrBG', 'BrBG_r', 'BuGn'
        plt.title(f"{class_labels[i]} (Total: {class_counts[class_labels[i]]})")
        plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
def plot_gallery(images, titles, h, w, n_row=3, n_col=3):
    """Helper function to plot a gallery of portraits"""
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        plt.title(titles[i], size=12)
        plt.xticks(())
        plt.yticks(())
```

```python
print("Fitting the classifier to the training set")
t0 = time()
param_grid = {'C': [1e3, 5e3, 1e4, 5e4, 1e5],
              'gamma': [0.0001, 0.0005, 0.001, 0.005, 0.01, 0.1], }

#class_weight='balanced'
clf = GridSearchCV(SVC(kernel='rbf', class_weight='balanced'), param_grid) #'precomputed', 'linear', 'sigmoid', 'poly', 'rbf'
clf = clf.fit(X_train_pca, y_train_)
print("done in %0.3fs" % (time() - t0))
print("Best estimator found by grid search:")
print(clf.best_estimator_)
```

```python
from sklearn.metrics import classification_report, accuracy_score

# Make predictions using the SVM model
y_pred = clf.predict(X_val_pca)

# Print classification report
y_val = np.argmax(y_validation, axis=-1)
print(classification_report(y_val, y_pred))

# Print training and validation accuracy
train_acc = clf.score(X_train_pca, np.argmax(y_train, axis=-1))  # Calculate training accuracy
val_acc = clf.score(X_val_pca, y_val)  # Calculate validation accuracy

print(f'Training Accuracy: {train_acc * 100:.2f}%')
print(f'Validation/Test Accuracy: {val_acc * 100:.2f}%')

# Calculate and print SVM accuracy
accuracy = accuracy_score(y_val, y_pred)
print(f'SVM Accuracy: {accuracy * 100:.2f}%')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout

# Define your model
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(X_train_pca.shape[1],)))
model.add(Dropout(0.5))
model.add(Dense(n_classes, activation='softmax'))  # Adjust n_classes based on your labels

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Fit the model and save the history
history = model.fit(X_train_pca, y_train, validation_data=(X_val_pca, y_validation), epochs=20, batch_size=32)
```

```python
import matplotlib.pyplot as plt

def plot_curves(history):
    # Get data from history
    epochs = range(1, len(history.history['loss']) + 1)
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    accuracy = history.history['accuracy']
    val_accuracy = history.history['val_accuracy']

    plt.figure(figsize=(12, 5), facecolor='white')  # Set figure size and background color

    # Plot loss
    ax1 = plt.subplot(1, 2, 1)
    ax1.set_facecolor('white')  # Set subplot background to white
    ax1.plot(epochs, loss, label="Training Loss", color='blue', linewidth=2)
    ax1.plot(epochs, val_loss, label="Validation Loss", color='orange', linewidth=2)
    ax1.title.set_text("Loss")
    ax1.set_xlabel("Epochs", fontsize=10)
    ax1.legend(fontsize=10)
    ax1.tick_params(axis='both', which='major', labelsize=8)  # Set font size for ticks

    # Add a thin black line box around the plot
    for spine in ax1.spines.values():
        spine.set_linewidth(0.5)  # Set spine width to 0.5
        spine.set_color('black')  # Set spine color to black

    # Plot accuracy
    ax2 = plt.subplot(1, 2, 2)
    ax2.set_facecolor('white')  # Set subplot background to white
    ax2.plot(epochs, accuracy, label="Training Accuracy", color='blue', linewidth=2)
    ax2.plot(epochs, val_accuracy, label="Validation Accuracy", color='orange', linewidth=2)
    ax2.title.set_text("Accuracy")
    ax2.set_xlabel("Epochs", fontsize=10)
    ax2.legend(fontsize=10)
    ax2.tick_params(axis='both', which='major', labelsize=8)  # Set font size for ticks

    # Add a thin black line box around the plot
    for spine in ax2.spines.values():
        spine.set_linewidth(0.5)  # Set spine width to 0.5
        spine.set_color('black')  # Set spine color to black

    plt.tight_layout()  # Adjust subplots to fit into figure area.
    plt.show()

# Call the function to plot the curves
plot_curves(history)  # Ensure to use the correct variable name for your history
```
